```python
import sys
from pathlib import Path

PROJECT_ROOT = Path().resolve().parent
sys.path.append(str(PROJECT_ROOT))

print("Added to path:", PROJECT_ROOT)

import torch
from torchvision.datasets import MNIST
from torchvision import transforms
import matplotlib.pyplot as plt

from src.utils.corruptions import (
    pixel_dropout,
    gaussian_noise,
    stripe,
    random_patch,
)

dataset = MNIST(
    root="data",
    train=True,
    download=True,
    transform=transforms.ToTensor(),
)

img, _ = dataset[0]

print(pixel_dropout)
print(gaussian_noise)
print(stripe)
print(random_patch)
```

```python
funcs = [pixel_dropout, gaussian_noise, stripe, random_patch]
titles = ["Pixel Dropout", "Gaussian Noise", "Stripe", "Random Patch"]

fig, axes = plt.subplots(1, 4, figsize=(12, 3))

for i, (fx, title) in enumerate(zip(funcs, titles)):
    out = fx(img)
    axes[i].imshow(out.squeeze(), cmap='gray')
    axes[i].set_title(title)
    axes[i].axis("off")

plt.show()
```
